````python
def add_numbers(num1: int | float, num2: int | float) -> int | float:
    """
    Description: THis function adds two numbers

    Args:
        - num1: [required] number1 (int or float)
        - num2: [required] number2 (int or float)

    Return:
        - a number num1 + num2 (int or float)
    """

    assert type(num1) == int or type(num1) == float, "The num1 should be either int or float type"
    assert type(num2) == int or type(num2) == float, "The num1 should be either int or float type"

    return num1 + num2
````

````python
add_numbers(3, 6.9)
````

````python
add_numbers('a', 3.3)
````

````python
# factorial of a postive integer

def factorial(n: int) -> int:
    """
    Description: This function will compute the factorial of a postive integer

    Input:
        - n: (int) The positive integer number whose factorial we want to compute

    Return:
        - (int) It returns a number which is factorial of the input number n
    """

    assert type(n) == int, "The value must be integer type"

    assert n > 0, "The value must be a positive integer"

    result = 1

    for i in range(2, n+1):
        result *= i

    return result
````

````python
factorial('25')
````

````python
factorial(6)
````

````python
factorial(8.)
````

**Binomial coefficient**

$$ \binom{n}{k} = \frac{n!}{k!(n-k)!};\ \ k <= n $$

````python
def binom_coeff(n: int,k: int) -> int:
    """
    docstring
    """

    # sanity check (using assert)

    result = factorial(n)/ (factorial(k) * factorial(n-k))

    return int(result)
````

````python
binom_coeff(6, 2)
````

### Python try, except (Exception handling)

```python
try:
    try block lets you test a block of code for errors
except <errortype-1>:
    this block lets you to handle the error for error type-1
except <errortype-2>:
    this block lets you handle the error for error type-2
else:
    this block will be executed when there is no error
finally:
    this block ill execute code, regardless of the try and except block
```

````python
print(abc)
````

````python
try:
    print(abc)
except NameError:
    print("The variable is not defined")
````

````python
try:
    x = 10
    z = 0
    print(x/z)
except NameError:
    print("The variable is not defined")
except Exception as e:
    print(f"Some other error occurred, error={e}")
````

````python
try:
    n = 8.
    v = factorial(n)
except Exception as e:
    print(f"Error = {e}")
else:
    print(f"factorial of {n} = {v}")
finally:
    print(f"The code execution ended")
````
