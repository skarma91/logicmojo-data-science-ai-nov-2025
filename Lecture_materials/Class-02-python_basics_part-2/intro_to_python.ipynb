## Printing the are of a triangle given the sides

The sides of triangle are a, b, c

$$semi-perimeter: s = 0.5*(a+b+c)$$

$$area=\sqrt{s(s-a)(s-b)(s-c)}$$

````python
a = float(input("enter side-1: "))
b = float(input("enter side-2: "))
c = float(input("enter side-3: "))

print("The sides are: ", a, b, c)
````

````python
s = 0.5*(a+b+c)

area = (s*(s-a)*(s-b)*(s-c))**0.5

print("The area is: ", area)
````

### Flow control

if-else statement

```python

if (condition):
    run this code block
else:
    run this code block
```

````python
my_name = "Akash"

if my_name == "Sourav":
    print("I am an instructor")
else:
    print("I am not an instructor")
````

````python
# define: magic number is a integer number which is divisble by both 5 and 7

number = input("Enter the number: ")
number = int(float(number))

if number % 5 == 0 and number % 7 == 0:
    print("The number is divisible by both 5 and 7")
    print("It is a magic number")
else:
    print("It is not a magic number")
````

### Multiple if-else (if-elif-else)

```python

if (condition-1):
    run this code block
elif (condition-2):
    run this block
elif (condition-3):
    run this block
else:
    run this block
```

````python
x = 5

if x > 20:
    print("The number is greater than 20")
elif x > 10:
    print("The number is greater than 10")

print("It is stopped")
````

### Nested if-else

```python
if (condition-1):
    if (condition-2):
        run this code block
    else:
        run this code bloc
else:
    run this code block
```

````python
x = 15

if (x > 10):
    if (x > 20):
        print("The number is both greater than 10 and 20")
    else:
        print("The number is between 10 and 20")
else:
    print("The number is less than or equal to 10")
````
